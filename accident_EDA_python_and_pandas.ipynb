```python
df.Start_Time = pd.to_datetime(df.Start_Time, errors='coerce')
df.Start_Time[0]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot histogram of hours with percentages
sns.histplot(df.Start_Time.dt.hour, bins=24, stat="percent")

plt.xlabel("Hour of the Day")
plt.ylabel("Percentage")
plt.title("Start Time Distribution by Hour (in %)")
plt.show()
```

. A high percentage of accident occur between 6am to 9am i think it probably lots of commerical business are getting there stuff to warehouses and people are hurry to get to work
. next highest percentage is between 3pm to 6pm

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Extract day of the week
df['Day_of_Week'] = df.Start_Time.dt.day_name()

# Set order of days
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Plot count of days
ax = sns.countplot(data=df, x='Day_of_Week', order=day_order)

# Convert y-axis to percentage
total = len(df)
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%'
    ax.annotate(percentage, (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom')

plt.xlabel("Day of the Week")
plt.ylabel("Count (with % shown above bars)")
plt.title("Start Time Distribution by Day of the Week")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The number of accident is less on weekend
is the distribution of accident by hour the same on weekend as on week days

## this graph shows the whole chat for sunday by hours

```python
df_sundays = df[df.Start_Time.dt.day_name() == 'Sunday']

# Extract the hour from Start_Time
sunday_hours = df_sundays.Start_Time.dt.hour

# Plot histogram of hours (only Sundays) with percentages
sns.histplot(sunday_hours, bins=12, stat="percent")

plt.xlabel("Hour of the Day")
plt.ylabel("Percentage")
plt.title("Start Time Distribution by Hour on Sundays (in %)")
plt.xticks(range(0, 24))  # Show all hours on x-axis
plt.tight_layout()
plt.show()
```

on sundays the peak occurs between 10 am to 3pm

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create ordered month category
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

months = df.Start_Time.dt.month_name()
months = pd.Categorical(months, categories=month_order, ordered=True)

# Plot with months in order
sns.displot(months, stat="percent", bins=12, height=6, aspect=2)

plt.xticks(rotation=45)
plt.title("Start Times by Month (%)")
plt.xlabel("Month")
plt.ylabel("Percentage")
plt.tight_layout()
plt.show()
```
